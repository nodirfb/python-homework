```python
import pandas as pd
```

```python
df = pd.read_csv('task\\sales_data.csv')
```

```python
# 1. Group the data by the `Category` column and calculate the following aggregate statistics for each category:
#    - Total quantity sold.
#    - Average price per unit.
#    - Maximum quantity sold in a single transaction.

df.groupby('Category').agg(total_quantity = ('Quantity','sum'),max_quantity = ('Quantity','max'),avg_price = ('Price','mean'))
df.head(10)
```

```python
# 2. Identify the top-selling product in each category based on the total quantity sold.
# df.groupby(['Product','Category'])['Quantity'].max()

df.groupby(['Category', 'Product'])['Quantity'].sum().reset_index() \
  .sort_values(['Category', 'Quantity'], ascending=[True, False]) \
  .drop_duplicates(subset=['Category'], keep='first')
```

```python
# 3. Find the date on which the highest total sales (quantity * price) occurred.
df.head(10)

df['total_sales'] = df['Quantity']*df['Price']
df.nlargest(1,'total_sales')
# df.sort_values('total_sales',ascending=False)
```
