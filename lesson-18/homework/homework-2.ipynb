```python
import pandas as pd
df = pd.read_csv('task\\tackoverflow_qa.csv')  # Double backslashes
df.head()
```

```python
# 1. Find all questions that were created before 2014

df['creationdate'] = pd.to_datetime(df['creationdate'],format='%Y-%m-%d %H:%M:%S')
df[df['creationdate'] < '2014']
```

```python
# 2. Find all questions with a score more than 50
df[df['score'] > 50]
```

```python
# 3. Find all questions with a score between 50 and 100
mask = df['score'].between(50,100)
df[mask]
```

```python
# 4. Find all questions answered by Scott Boston
mask = df['ans_name'] == 'Scott Boston'
df[mask]
```

```python
# 5. Find all questions answered by the following 5 users

# ?? this task asks me to find questions by following 5 user. but in this task there are not given users.

mask = df['ans_name'].isin(['Adam Hughes','Avaris','doug','unutbu','joris'])
df[mask]
```

```python
# 6. Find all questions that were created between March, 2014 and October 2014 that were answered by Unutbu and have score less than 5.
mask = (df['creationdate'].between('2014-03','2014-10')) & (df['ans_name'] == 'unutbu') & (df['score'] < 5)
df[mask] # 
```

```python
# 7. Find all questions that have score between 5 and 10 or have a view count of greater than 10,000
mask = (df['score'].between(5,10)) | (df['viewcount'] > 10000)
df[mask]
```

```python
# 8. Find all questions that are not answered by Scott Boston
mask = df['ans_name'] == 'Scott Boston'
df[~mask]
# df.info()
```
